# Splitting the data set into independent (X) and dependent (Y) data sets

```python
X = df.iloc[:,2:31].values
Y = df.iloc[:,1].values
```

# Splitting the data set into 75% training and 25% testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test  = sc.transform(X_test)
```

```python
X_train
```

# Create a function for the models

```python
def models(X_train,Y_train):
    #Logistic Regression
    from sklearn.linear_model import LogisticRegression 
    log = LogisticRegression(random_state=0)
    log.fit(X_train,Y_train)
    
    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion='entropy',random_state=0)
    tree.fit(X_train,Y_train)
    
    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
    forest.fit(X_train,Y_train)
    
    #Print the output of training the 3 models
    print("Regression model result " , log.score(X_train,Y_train))
    print("Decision Tree classifier result " , log.score(X_train,Y_train))
    print("Random Forest cassifier result " , log.score(X_train,Y_train))
    
    return log , tree, forest
    
    
    
```

```python
model = models(X_train,Y_train)
```

# Testing the model on test data and using the confusion matrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,model[0].predict(X_test))

TP = cm[0][0]
TN = cm[1][1]
FP = cm[0][1]
FN = cm[1][0]

print(cm)

print('testing accuracy of logistic regression model is equal to : ', (TP+TN)/(TP+TN+FP+FN))
```

```python
from sklearn.metrics import confusion_matrix

for i in range(len(model)):
    cm = confusion_matrix(Y_test,model[i].predict(X_test))
    
    TP = cm[0][0]
    TN = cm[1][1]
    FP = cm[0][1]
    FN = cm[1][0]
    
    print(cm)
    print('testing accuracy of logistic regression model is equal to : ', (TP+TN)/(TP+TN+FP+FN))
```
